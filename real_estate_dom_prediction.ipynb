```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
from collections import Counter
from scipy.stats.stats import pearsonr
```

```python
np.set_printoptions(precision=4, suppress=True)
%matplotlib inline
rcParams['figure.figsize'] = 10, 10
plt.style.use('seaborn-whitegrid')
```

```python
dataset = pd.read_csv('properties.csv')
dataset.columns = ['zip','price','sq_ft','lot_size','beds','baths','age','daysOnMarket']
```

```python
dataset.shape
```

```python
dataset = dataset[dataset['baths'] <= 6]
dataset = dataset[dataset['daysOnMarket'] <= 730]
dataset = dataset[dataset['zip'] <= 93000]
dataset = dataset[dataset['lot_size'] < 1]
dataset = dataset[dataset['price'] < 2000000]
dataset = dataset[dataset['age'] > 1]
```

```python
sb.pairplot(dataset)
```
